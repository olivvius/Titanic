```python
import pandas as pd
import statistics as st
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix
train=pd.read_csv('train.csv',sep=',')
train.set_index('PassengerId',inplace=True,drop=True)
print(train.columns)
train.dtypes
print(train.size)
print(train.ndim)
def parse_model_0(DF):
    target=DF.Survived
    DF2=DF[['Fare','SibSp','Parch']]
#    DF2=DF.take([8,5,6], axis=1)
    return DF2, target
X,y=parse_model_0(train)
print(X,y)

from sklearn.model_selection import train_test_split
#model = LogisticRegression(solver='liblinear', random_state=0).fit(X, y)
#print(model.predict_proba(X))
#print(model.score(X, y))

from sklearn.model_selection import cross_val_score

def compute_score(clf, X, y):
    xval = cross_val_score(clf, X, y, cv = 8)
    return np.mean(xval)

from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(max_iter=5000)
print("score of logreg using crossvalidations : {:.2f} ".format(compute_score(lr, X, y)))

train.head(10)
```

```python
def parse_model_1(DF):
    target2 = DF.Survived
    for dum in DF['Pclass']:
        class_dummies = pd.get_dummies(DF['Pclass'],prefix='Class')
        DF2 = DF.join(class_dummies)
    for dum in DF2['Sex']:
        sex_dummies = pd.get_dummies(DF2['Sex'],prefix='Sex')
        DF3 = DF2.join(sex_dummies)
    to_del = ['Name', 'Cabin', 'Embarked', 'Survived','Ticket','Pclass','Sex','Age']
    for col in to_del : del DF3[col]
    return DF3, target2
X2,y=parse_model_1(train)
print(X2)
lr.fit(X2, y)
print("score of logreg using crossvalidations : {:.3f} ".format(compute_score(lr, X2, y)))
print("Coef of each variables : {}".format(lr.coef_))
```
